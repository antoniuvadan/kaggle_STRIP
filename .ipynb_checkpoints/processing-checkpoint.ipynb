# Preprocessing

Preprocessing consists of the following steps:
- pad image with zeros to make dimensions multiples of TILE_SIZE
- decompose image into TILE_SIZE x TILE_SIZE tiles
- downsample each tile by DOWNSAMPLE_FACTOR
- convert tiles to grayscale
- set background colour to white -- some images have a light gray background
- delete tiles that are predominantly white

```python
def white_background_gs(img_gray: np.ndarray, threshold: float) -> np.ndarray:
    """
    Returns array with values greater than threshold set to 1.
    
    img_gray: grayscale image with values normalized from 0 to 1.
    threshold: value between 0 and 1.
    """
    img_gray[img_gray > threshold] = 1
    return img_gray

def white_background_rgb(img_rgb: np.ndarray, threshold: int) -> np.ndarray:
    """
    Returns array with values greater than threshold set to [255,255,255].
    
    img_rgb: rgb image with values between 0 and 255.
    threshold: value between 0 and 255*3=765.
    """
    img_rgb[np.sum(img_rgb, axis=2) > threshold] = [255,255,255]
    return img_rgb
```

```python
def pad(img_gray: np.ndarray, tile_size: int) -> np.ndarray:
    """
    Pads img with 1s at the edges such that the new array's dimensions are multiples of
    tile_size.
    """
    height, width = img_gray.shape[:2]
    pad_h = tile_size - (height % tile_size)
    pad_w = tile_size - (width % tile_size)

    # Compute padding for each dimension.
    # Padding is split equally along both directions on each axis.
    pad_h_0 = 0
    pad_h_1 = 0
    pad_w_0 = 0
    pad_w_1 = 0

    if pad_h & 1 == 0:
        pad_h_0 = int(pad_h / 2)
        pad_h_1 = int(pad_h / 2)
    else:
        pad_h_0 = int(pad_h / 2)
        pad_h_1 = int(pad_h / 2) + 1

    if pad_w & 1 == 0:
        pad_w_0 = int(pad_w / 2)
        pad_w_1 = int(pad_w / 2)
    else:
        pad_w_0 = int(pad_w / 2)
        pad_w_1 = int(pad_w / 2) + 1
    
    return np.pad(img_gray, ((pad_h_0,pad_h_1), (pad_w_0,pad_w_1)), mode='constant', 
                  constant_values=1)

def pad_rgb(img_rgb: np.ndarray, tile_size: int) -> np.ndarray:
    """
    Pads img with (255,255,255) at the edges such that the new array's dimensions are 
    multiples of tile_size.
    """
    height, width = img_rgb.shape[:2]
    pad_h = tile_size - (height % tile_size)
    pad_w = tile_size - (width % tile_size)

    # Compute padding for each dimension.
    # Padding is split equally along both directions on each axis.
    pad_h_0 = 0
    pad_h_1 = 0
    pad_w_0 = 0
    pad_w_1 = 0

    if pad_h & 1 == 0:
        pad_h_0 = int(pad_h / 2)
        pad_h_1 = int(pad_h / 2)
    else:
        pad_h_0 = int(pad_h / 2)
        pad_h_1 = int(pad_h / 2) + 1

    if pad_w & 1 == 0:
        pad_w_0 = int(pad_w / 2)
        pad_w_1 = int(pad_w / 2)
    else:
        pad_w_0 = int(pad_w / 2)
        pad_w_1 = int(pad_w / 2) + 1
    
    return np.stack([np.pad(img_rgb[:,:,c], ((pad_h_0,pad_h_1), (pad_w_0,pad_w_1)), mode='constant', constant_values=255) for c in range(3)], axis=2)
```

```python
def valid_tile_gs(tile: np.ndarray, side_len: int, tile_valid_threshold: float) -> bool:
    """
    tile:                 greyscale image with values normalized to the range 0 - 1.
    side_len:             side length of tile.
    tile_valid_threshold: maximum allowed proportion of white tile pixels.

    Returns: True if tile contains sufficient values not equal to 1.
             False otherwise.
    """
    if np.count_nonzero(tile == 1) > (side_len ** 2) * tile_valid_threshold:
        return False
    return True


def valid_tile_rgb(tile: np.ndarray, side_len: int, tile_valid_threshold: float) -> bool:
    """
    tile:                 rgb image with values in the range 0-255.
    side_len:             side length of tile.
    tile_valid_threshold: maximum allowed proportion of white tile pixels.

    Returns: True if tile contains sufficient pixels not equal to [255,255,255].
             False otherwise.
    """
    if np.count_nonzero(np.all(tile == [255,255,255], axis=2)) > (side_len ** 2) * \
        tile_valid_threshold:
        return False
    return True
```

```python
def rescale(img: np.ndarray, downsample_factor: int) -> np.ndarray:
    """
    Samples every `scale_factor` pixels from img.

    downsample_factor: factor by which original image is downsampled.

    Returns: Downsampled array.
    """
    return img[::downsample_factor, ::downsample_factor]
```

```python
def generate_tiles_gs(img_rgb: np.ndarray, tile_side_len: int, downsample_factor: int, 
                       tile_valid_threshold: float, img_name: str, dir_path: str) -> None:
    """
    Decomposes img_rgb down into tiles, downsamples using given ratio, and saves 
    the grayscale tile at specified dir_path in a new directory named `img_name`. 
    Each file will be named `img_name_tile_<tile_i>_<tile_j>`.

    img_rgb:              rgb image with values 0-255 along each color channel.
    tile_side_len:        side length of generated tiles.
    downsample_factor:    factor by which original image is downsampled.
    tile_valid_threshold: maximum allowed proportion of white tile pixels.
    img_name:             name of img_rgb.
    dir_path:             directory path where a new directory called `img_name` will
                          be created with tile contents. Path does not end with `/`
                          character.

    Returns: None
    """
    tiles_dir_path = dir_path + '/' + img_name + '_' + str(tile_side_len) + \
                     '_' + str(downsample_factor) + '_' + str(int(tile_valid_threshold*100)) 
    Path(tiles_dir_path).mkdir(parents=True, exist_ok=True)
    
    height, width = img_rgb.shape[:2]
    for i in range(0, height, tile_side_len):
        for j in range(0, width, tile_side_len):
            tile = img_rgb[i:i+tile_side_len, j:j+tile_side_len]
            tile_scaled = rescale(tile, downsample_factor);
            tile_scaled_gray = color.rgb2gray(tile_scaled)
            tile_scaled_gray_wbg = white_background_gs(tile_scaled_gray, WHITE_THRESHOLD)

            if not valid_tile_gs(tile_scaled_gray_wbg, tile_scaled_gray_wbg.shape[0], 
                              tile_valid_threshold):
                continue

            tile_filepath = tiles_dir_path + '/' + img_name + '_tile_' + str(i) + '_' + str(j) + \
                            '.png'
            PIL.Image.fromarray((tile_scaled_gray_wbg * 255).astype(np.uint8)).save(tile_filepath)
            
def generate_tiles_rgb(img_rgb: np.ndarray, tile_side_len: int, downsample_factor: int,
                       white_threshold: int, tile_valid_threshold: float, img_name: str, 
                       dir_path: str) -> None:
    """
    Decomposes img_rgb down into tiles, downsamples using given ratio, and saves 
    the rgb tile at specified dir_path in a new directory named `img_name`. 
    Each file will be named `img_name_tile_<tile_i>_<tile_j>`.

    img_rgb:              rgb image with values 0-255 along each color channel.
    tile_side_len:        side length of generated tiles.
    downsample_factor:    factor by which original image is downsampled.
    white_threshold:      maximum valid sum of RGB channels. Pixels with sums above
                          this threshold are set to white. 
                          0 <= white_threshold <= 765
    tile_valid_threshold: maximum allowed proportion of white tile pixels.
    img_name:             name of img_rgb.
    dir_path:             directory path where a new directory called `img_name` will
                          be created with tile contents. Path does not end with `/`
                          character.

    Returns: None
    """
    tiles_dir_path = dir_path + '/' + img_name + '_' + str(tile_side_len) + \
                     '_' + str(downsample_factor) + '_' + str(int(tile_valid_threshold*100)) 
    Path(tiles_dir_path).mkdir(parents=True, exist_ok=True)

    height, width = img_rgb.shape[:2]
    for i in range(0, height, tile_side_len):
        for j in range(0, width, tile_side_len):
            tile = img_rgb[i:i+tile_side_len, j:j+tile_side_len]
            tile_scaled = rescale(tile, downsample_factor);
            tile_scaled_wbg = white_background_rgb(tile_scaled, white_threshold)
            tile_scaled_sidelen = tile_scaled_wbg.shape[0]

            if not valid_tile_rgb(tile_scaled_wbg, tile_scaled_sidelen, tile_valid_threshold):
                continue

            tile_filepath = tiles_dir_path + '/' + img_name + '_tile_' + str(i) + '_' + str(j) + \
                            '.png'
            PIL.Image.fromarray(tile_scaled_wbg).save(tile_filepath)
            
def generate_tiles(img_rgb: np.ndarray, tile_side_len: int, downsample_factor: int, 
                   white_threshold_gs: float, tile_valid_threshold: float, img_name: str, 
                   dir_path: str) -> None:
    """
    Decomposes img_rgb down into tiles, downsamples using given ratio, and saves 
    the grayscale as well as rgb tiles at the specified dir_path in new directories
    named `img_name`/gs and `img_name`/rgb.
    Each tile image file will be named `tile_<tile_i>_<tile_j>`.

    img_rgb:              rgb image with values 0-255 along each color channel.
    tile_side_len:        side length of generated tiles.
    downsample_factor:    factor by which original image is downsampled.
    white_threshold_gs:   maximum valid grayscale pixel value. Pixels with values above
                          this threshold are set to white.
                          0 <= white_threshold <= 1
    tile_valid_threshold: maximum allowed proportion of white tile pixels.
    img_name:             name of img_rgb.
    dir_path:             directory path where a new directory called `img_name` will
                          be created with tile contents. Path does not end with `/`
                          character.

    Returns: None
    """
    tiles_rgb_dir_path_valid = dir_path + '/' + img_name + '_' + str(tile_side_len) + \
                 '_' + str(downsample_factor) + '_' + str(int(tile_valid_threshold*100)) + \
                 '/rgb/valid'
    tiles_rgb_dir_path_invalid = dir_path + '/' + img_name + '_' + str(tile_side_len) + \
                 '_' + str(downsample_factor) + '_' + str(int(tile_valid_threshold*100)) + \
                 '/rgb/invalid'
    tiles_gs_dir_path_valid = dir_path + '/' + img_name + '_' + str(tile_side_len) + \
                 '_' + str(downsample_factor) + '_' + str(int(tile_valid_threshold*100)) + \
                 '/gs/valid'
    tiles_gs_dir_path_invalid = dir_path + '/' + img_name + '_' + str(tile_side_len) + \
                 '_' + str(downsample_factor) + '_' + str(int(tile_valid_threshold*100)) + \
                 '/gs/invalid'
    Path(tiles_rgb_dir_path_valid).mkdir(parents=True, exist_ok=True)
    Path(tiles_rgb_dir_path_invalid).mkdir(parents=True, exist_ok=True)
    Path(tiles_gs_dir_path_valid).mkdir(parents=True, exist_ok=True)
    Path(tiles_gs_dir_path_invalid).mkdir(parents=True, exist_ok=True)
    
    height, width = img_rgb.shape[:2]
    for i in range(0, height, tile_side_len):
        for j in range(0, width, tile_side_len):
            tile_rgb = img_rgb[i:i+tile_side_len, j:j+tile_side_len]
            tile_scaled_rgb = rescale(tile_rgb, downsample_factor);
            tile_scaled_gs = color.rgb2gray(tile_scaled_rgb)
            tile_scaled_gs_wbg = white_background_gs(tile_scaled_gs, white_threshold_gs)            

            # use tile_scaled_gray_wbg to mask tile_scaled_rgb            
            tile_scaled_rgb[tile_scaled_gs_wbg == 1] = [255,255,255]

            if not valid_tile_gs(tile_scaled_gs_wbg, tile_scaled_gs_wbg.shape[0], \
                              tile_valid_threshold):
                tile_filepath_rgb = tiles_rgb_dir_path_invalid + '/' + 'tile_' + \
                                    str(i) + '_' + str(j) + '.png'
                tile_filepath_gs = tiles_gs_dir_path_invalid + '/' + 'tile_' + \
                                    str(i) + '_' + str(j) + '.png'
                PIL.Image.fromarray(tile_scaled_rgb).save(tile_filepath_rgb)
                PIL.Image.fromarray((tile_scaled_gs_wbg * 255).astype(np.uint8))\
                   .save(tile_filepath_gs)
            else:
                tile_filepath_rgb = tiles_rgb_dir_path_valid + '/' + 'tile_' + \
                        str(i) + '_' + str(j) + '.png'
                tile_filepath_gs = tiles_gs_dir_path_valid + '/' + 'tile_' + \
                                    str(i) + '_' + str(j) + '.png'
                PIL.Image.fromarray(tile_scaled_rgb).save(tile_filepath_rgb)
                PIL.Image.fromarray((tile_scaled_gs_wbg * 255).astype(np.uint8))\
                   .save(tile_filepath_gs)
```

```python
# Generate tiles
train_set_dir = '../dataset_full/train'
tiles_dir_path = '/Volumes/Extreme SSD/cmpt400/tiles_prod'
num_files = len(os.listdir(train_set_dir))
for i, file in enumerate(os.listdir(train_set_dir)):
    print(str(i+1) + '/' + str(num_files) + ': ' + file)
    img_name = file[0:-4] # ignore .tif
    print('reading image...')
    img = tif.imread(train_set_dir + '/' + file)
    print('padding...')
    img_padded = pad_rgb(img, TILE_SIZE)
    print('generating tiles...')
    generate_tiles(img_padded, TILE_SIZE, DOWNSAMPLE_FACTOR, WHITE_THRESHOLD_GS, 
                       TILE_VALIDATION_THRESHOLD, img_name, tiles_dir_path)
    print()
    break

print('done')
```

```python
# Generate full grayscale image from tiles

tiles_path = '/Volumes/Extreme SSD/cmpt400/tiles_prod'
for i, file in enumerate(os.listdir(tiles_path)):
    gs_path = file + '/gs'
    rgb_path = file + '/rgb'
    print(str(i) + ': ' + file)
    # get all tile names and sort them by their x and y coordinates
    invalid_tiles = enumerate(os.listdir(tiles_path + '/' + gs_path + '/invalid'))
    invalid_tiles = [['/invalid', f[1], int(int(f[1].split('_')[1])/4096), int(int(f[1].split('_')[2].split('.')[0])/4096)] for f in invalid_tiles]
    valid_tiles = enumerate(os.listdir(tiles_path + '/' + gs_path + '/valid'))
    valid_tiles = [['/valid', f[1], int(int(f[1].split('_')[1])/4096), int(int(f[1].split('_')[2].split('.')[0])/4096)] for f in valid_tiles]
    all_tiles = sorted(invalid_tiles + valid_tiles, key=itemgetter(2,3))
    img_h = (all_tiles[-1][2] + 1) * TILE_W_H
    img_w = (all_tiles[-1][3] + 1) * TILE_W_H
    img_full = np.empty((img_h, img_w), dtype=np.float64)

    for t in all_tiles:
        t_path = tiles_path + '/' + gs_path + t[0] + '/' + t[1]
        tile = io.imread(t_path, as_gray=True)
        tile = tile/255
        y = t[2]
        x = t[3]
        img_full[y * TILE_W_H : (y+1) * TILE_W_H, x * TILE_W_H : (x+1) * TILE_W_H] = tile

    tile_filepath_gs = tiles_path + '/' + file + '/' + 'full_gs.png'
    PIL.Image.fromarray((img_full * 255).astype(np.uint8))\
                   .save(tile_filepath_gs)
print('done')
```

### Processing stages for HoC:
- close -- to smooth out the edges
- open  -- to remove small foreground pixel regions (many images have small "dots" 
  scattered across image)
- fill  -- to fill in gaps

```python
# Convert full grayscale images to binary, do a closing, opening, and fill.
tiles_path = '/Volumes/Extreme SSD/cmpt400/tiles_prod'
for i, file in enumerate(os.listdir(tiles_path)):
    print(str(i + 1) + ': ' + file)
    if file[0:4] == 'hocs':
        continue
    img_name = tiles_path + '/' + file + '/full_gs.png'
    img_gs = io.imread(img_name, as_gray=True)
    img_gs = img_gs/255
    img_gs[img_gs < 1] = 0 # convert to binary
    img_gs = np.logical_not(img_gs).astype(np.uint8) # flip bits

    closed = morph.binary_closing(img_gs, footprint=morph.disk(4))
    opened = morph.binary_opening(closed, footprint=morph.disk(4))
    filled = nd.binary_fill_holes(opened)
    
    # plt.imshow(filled, cmap="gray", vmin=0, vmax=1)
    # plt.show()

    file_dest = tiles_path + '/' + file + '/full_binary_processed.png'
    PIL.Image.fromarray((filled * 255).astype(np.uint8))\
                       .save(file_dest)
print('done')
```

# Create Train/Test sets from Kaggle's Train set
Due to not having checked the contents of the test directory and test.csv 
at the beginning of the project, I realized very late (March 11) that
there are only 4 images in the provided test set.
Below is the code that grabs 100 images from the train set and creates
a test set. Additionally creates a new test directory containing CE & LAA
subdirectories. The ratio of CE:LAA images is preserved in the test set.

```python
import numpy as np
import pandas as pd

# pd.set_option('display.max_columns', None)

print('starting program')
train_csv_path = '/Volumes/Extreme SSD/cmpt400/tiles_prod_csv/train.csv'
train_df = pd.read_csv(train_csv_path)
print('finished reading csv')
CE_count = len(train_df[train_df['label']=='CE'])
LAA_count = len(train_df[train_df['label']=='LAA'])
CE_ratio = CE_count / (CE_count + LAA_count)
LAA_ratio = LAA_count / (CE_count + LAA_count)
CE_test_count = int(CE_ratio * 95) # 95 is 15% of 632, the total number of patients
LAA_test_count = int(LAA_ratio * 95) + 1 # python truncates instead of
                                          # rounding

# print(train_df.iloc[0:10, :])
print(CE_count, CE_ratio, CE_test_count)
print(LAA_count, LAA_ratio, LAA_test_count)

patient_names = np.unique(train_df['patient_id'].to_numpy(copy=True))
np.random.shuffle(patient_names) # np.unique returns the unique elements sorted

print('number of patients:', len(patient_names))

test_CE = []
test_LAA = []
current_CE_count = 0
current_LAA_count = 0
print('iterating through images...')
for i, patient in enumerate(patient_names):
    label = train_df[train_df['patient_id'] == patient]['label'].to_numpy(copy=True)[0]
    if label == 'CE' and current_CE_count < CE_test_count:
        test_CE += [patient]
        current_CE_count += 1
    elif label and current_LAA_count < LAA_test_count:
        test_LAA += [patient]
        current_LAA_count += 1
print('test_CE', len(test_CE))
print(test_CE)
print('test_LAA', len(test_LAA))
print(test_LAA)
        
```
